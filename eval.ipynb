```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset list: (file path, display name)
datasets = [
    ("labeled_address_claim.csv", "Address Claim Attack"),
    ("labeled_dos.csv", "FMS Attack"),
    ("labeled_rq_overload.csv", "PGN Spam Attack"),
    ("labeled_TSC1.csv", "RPM Attack")
]

# Store counts
attack_counts = []
normal_counts = []

for file, _ in datasets:
    df = pd.read_csv(file)
    normal_count = (df['label'].str.lower() == 'normal').sum()
    attack_count = len(df) - normal_count

    normal_counts.append(normal_count)
    attack_counts.append(attack_count)

# Bar chart setup
labels = [name for _, name in datasets]
x = range(len(datasets))

plt.figure(figsize=(10, 6))
bar_width = 0.35

plt.bar(x, normal_counts, width=bar_width, label="Normal", color="green")
plt.bar([i + bar_width for i in x], attack_counts, width=bar_width, label="Attack", color="red")

plt.xlabel("Dataset")
plt.ylabel("Number of Messages")
plt.title("Normal vs. Attack Messages per Dataset")
plt.xticks([i + bar_width/2 for i in x], labels, rotation=20)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig("normal_vs_attack_counts.png", dpi=300)
plt.show()
```
